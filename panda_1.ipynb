```python
import pandas as pd
# loc (linhas , colunas )
# Criando um novo DataFrame
data = {'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
        'Idade': [25, 30, 35, 40],
        'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador']}

df = pd.DataFrame(data)

print("DataFrame Original:")
print(df)
linha_0 = df.loc[0]
print("Acessando a primeira linha:")
print(linha_0)

coluna_nome = df.loc[:, 'Nome']
print("Acessando a coluna 'Nome':")
print(coluna_nome)

valor_2_nome = df.loc[2, 'Nome']
print("Acessando o valor na linha 2 e coluna 'Nome':")
print(valor_2_nome)
```

```python
import pandas as pd

# Criando um DataFrame de exemplo
df = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [5, 6, 7, 8]},
                   index=['a', 'b', 'c', 'd'])

# Acessando a linha com rótulo 'b'
linha_b = df.loc['b'] # como loc é linha coluna , e so tem um parametro , entçao ele pegou tudo da linnha que tinha o termo b
print(linha_b) 
# Acessando a coluna 'A'
coluna_A = df.loc[:, 'A']

print(coluna_A)


```
