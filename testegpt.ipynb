```python
import numpy as np
import pandas as pd

# Criando um DataFrame de exemplo com informações de vendas
data = {'Vendedor': ['João', 'Maria', 'Pedro', 'Ana', 'Maria', 'João'],
        'Produto': ['A', 'B', 'A', 'C', 'B', 'C'],
        'Quantidade': [10, 5, 8, 12, 6, 15],
        'Preço_Unitário': [20, 30, 25, 15, 35, 20]}

df = pd.DataFrame(data)

# Calculando o total de vendas para cada vendedor usando NumPy
total_vendas = df.groupby('Vendedor')['Quantidade'].sum()

# Adicionando os totais de vendas como uma nova coluna no DataFrame
df['Total_Vendas'] = df['Vendedor'].map(total_vendas)

print(df)
```

```python
import pandas as pd
import openpyxl as op
from tqdm import tqdm

# Carregar a planilha Excel e o arquivo de processamento usando pandas
planilha = op.load_workbook(r'C:\Users\FARIAS\Documents\3. Danyel\Base teste pandas.xlsx')
Arquivo_processamento = pd.read_excel(r"C:\Users\FARIAS\Documents\3. Danyel\Base teste pandas2.xlsx")

# Acessar a aba ativa da planilha
aba = planilha.active

# Obter o número máximo de linhas na planilha
maximo_linha = aba.max_row

# Inicializar uma barra de progresso com o número total de iterações
progress_bar = tqdm(total=len(Arquivo_processamento), desc="Processando linhas")

# Iterar sobre as linhas do DataFrame e atribuir os valores às células correspondentes na planilha
for index, row in Arquivo_processamento.iterrows():
    for col, value in enumerate(row, start=1):
        aba.cell(row=maximo_linha + index + 1, column=col, value=value)
    progress_bar.update(1)  # Atualizar a barra de progresso para indicar a conclusão de uma iteração

# Finalizar a barra de progresso
progress_bar.close()

# Salvar a planilha
planilha.save("TSTES.xlsx")

print(maximo_linha)



```

```python
import requests

# URL base do site do SharePoint
base_url = "https://escolatrabalhador4.sharepoint.com/sites/Teste_123"

# ID da lista (se disponível)
list_id = "a365f173-34a5-409e-865c-da68f83ab518"

# Endpoint para obter os itens da lista
endpoint = f"/_api/web/lists('{list_id}')/items"

# Dados de autenticação (se necessário)
auth = ("seu_usuario", "sua_senha")

# Headers necessários para a solicitação
headers = {
    "Accept": "application/json;odata=verbose"
}

# Função para obter os itens da lista do SharePoint
def get_list_items():
    try:
        # Fazendo uma solicitação GET para obter os itens da lista
        response = requests.get(base_url + endpoint, headers=headers, auth=auth)
        if response.status_code == 200:
            # Se a solicitação for bem-sucedida, retornamos os dados
            return response.json()
        else:
            print("Erro ao obter itens da lista:", response.text)
            return None
    except Exception as e:
        print("Erro na solicitação:", e)
        return None

# Obtendo os itens da lista do SharePoint
lista_itens = get_list_items()
if lista_itens:
    # Processando os itens obtidos
    for item in lista_itens["d"]["results"]:
        print(item)  # Aqui você pode processar cada item conforme necessário
else:
    print("Não foi possível obter os itens da lista do SharePoint.")
```
