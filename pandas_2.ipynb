```python
import xml.etree.ElementTree as ET

# Carregar o arquivo XML
tree = ET.parse(r'C:\Users\FARIAS\Downloads\base xml\22230910854165001741550010000206641896537151-nfe.xml')
root = tree.getroot()

# Encontrar e iterar sobre os elementos 'det'
for det in root.findall('.//{http://www.portalfiscal.inf.br/nfe}det'):
    nItem = det.attrib['nItem']
    xProd = det.find('.//{http://www.portalfiscal.inf.br/nfe}xProd').text
    vProd = det.find('.//{http://www.portalfiscal.inf.br/nfe}vProd').text
    print(f"Item {nItem}: {xProd} - R$ {vProd}")

# Encontrar e exibir informações do total
total = root.find('.//{http://www.portalfiscal.inf.br/nfe}total')
vNF = total.find('.//{http://www.portalfiscal.inf.br/nfe}vNF').text
print(f"Total NF-e: R$ {vNF}")
```
