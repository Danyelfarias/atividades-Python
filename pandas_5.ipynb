```python
import pandas as pd
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
caminho = r"C:\Users\FARIAS\Documents\3. Danyel\Base teste pandas2.xlsx"

BasePrincipal = pd.read_excel(caminho,engine="openpyxl")
# BasePrincipal
BasePrincipal = BasePrincipal[['ID Cliente','Data','ID Pedido','Categoria ','Valor']]
Coluna_Mes =  pd.to_datetime(BasePrincipal['Data']) # converter para data a coluna
BasePrincipal['Nome do Mês']=BasePrincipal['Data'].dt.strftime('%B')
BasePrincipal
```

```python

BasePrincipal.loc[BasePrincipal['Valor']<2000,'Faixa Valor']='baixo'

BasePrincipal.loc[(BasePrincipal['Valor']>2000) & (BasePrincipal['Valor']<4000) ,'Faixa Valor']='medio'
BasePrincipal.loc[BasePrincipal['Valor']>5000,'Faixa Valor']= 'alto'
BasePrincipal 
```

```python
ordem_meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
               'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']

BasePrincipal['Nome do Mês'] = pd.Categorical(BasePrincipal['Nome do Mês'], categories=ordem_meses, ordered=True)
resultado_groupby = BasePrincipal.groupby(['Categoria ','Nome do Mês'])['Valor'].sum()
resultado_df = resultado_groupby.reset_index()
resultado_df
```
