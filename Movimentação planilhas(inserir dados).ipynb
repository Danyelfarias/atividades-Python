```python
import openpyxl
import matplotlib.pyplot as plt

# Passo 1: Carregar o arquivo Excel
arquivo_excel = "CADASTRO DE PRODUTO.xlsx"
workbook = openpyxl.load_workbook(arquivo_excel)
sheet = workbook.active

# Passo 2: Extrair os dados das colunas relevantes
produtos = []
quantidades = []

for row in sheet.iter_rows(min_row=2, max_row=sheet.max_row, min_col=1, max_col=2, values_only=True):
    produto, quantidade = row
    produtos.append(produto)
    quantidades.append(quantidade)

# Passo 3: Criar o gráfico de barras
plt.figure(figsize=(10, 6))  # Define o tamanho do gráfico
plt.bar(produtos, quantidades, color='skyblue')

# Adiciona rótulos e título
plt.xlabel('Produtos')
plt.ylabel('Quantidade')
plt.title('Quantidade de cada Produto')

# Rotaciona os rótulos no eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Exibe o gráfico
plt.tight_layout()
plt.show()
```
