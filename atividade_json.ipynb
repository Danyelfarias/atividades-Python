```python
import requests
import pandas as pd

# URL da API
url = 'https://servicodados.ibge.gov.br/api/v1/localidades/estados/23/distritos?orderBy=nome'

# Requisição GET para obter os dados
response = requests.get(url)

# Verifica se a requisição foi bem sucedida
if response.status_code == 200:
    # Carrega os dados da resposta em JSON
    dados = response.json()

    # Cria um DataFrame vazio
    df = pd.DataFrame()

    # Itera sobre os distritos obtidos na resposta
    for distrito in dados:
        # Converte o dicionário do distrito em um DataFrame temporário
        df_temp = pd.json_normalize(distrito)
        
        # Concatena o DataFrame temporário ao DataFrame principal
        df = pd.concat([df, df_temp], ignore_index=True)

    # Exibe o DataFrame com todas as informações possíveis
    display(df.head())
else:
    print('Falha ao obter os dados. Código de status:', response.status_code)
```
